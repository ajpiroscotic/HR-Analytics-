```python
df_num = df1[['Age','DailyRate','DistanceFromHome', 
                       'EnvironmentSatisfaction', 'HourlyRate',                     
                       'JobInvolvement', 'JobLevel',
                       'JobSatisfaction', 
                       'RelationshipSatisfaction', 
                       'StockOptionLevel',
                        'TrainingTimesLastYear']].copy()
```

```python
df_num.head()
```

```python
df_num.describe()
```

```python
df_cat = df1[['Attrition', 'BusinessTravel','Department',
                       'EducationField','Gender','JobRole',
                       'MaritalStatus',
                       'Over18', 'OverTime']].copy()
```

```python
df_cat.head()
```

```python
df_cat['BusinessTravel'].value_counts()
```

```python
df_cat['Department'].value_counts()
```

```python
df_cat['EducationField'].value_counts()
```

```python
df_cat['JobRole'].value_counts()
```

```python
new_values={'Yes':1,'No':0}
df_cat['Attrition']=df_cat['Attrition'].map(new_values)
```

```python
df_cat.head()
```

```python
df_cat=pd.get_dummies(df_cat)
```

```python
df_cat.head()
```

```python
df=pd.concat([df_cat,df_num],axis=1)
df.head()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x = df['Attrition']
y = df.drop('Attrition', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(y, x, test_size=0.4, random_state=10)
```

```python
randfc = RandomForestClassifier()
randfc.fit(X_train,y_train)
```

```python
rfc_pred = randfc.predict(X_test)
```

```python
accuracy_score(y_test, rfc_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
knn_pred=knn.predict(X_test)
```

```python
accuracy_score(y_test, knn_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gradb = GradientBoostingClassifier(n_estimators=100)
```

```python
gradb.fit(X_train,y_train)
```

```python
gradb_pred=gradb.predict(X_test)
```

```python
accuracy_score(y_test,gradb_pred)
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
```

```python
X_train_norm=norm.fit_transform(X_train)
```

```python
X_test_norm=norm.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train_norm,y_train)
```

```python
lr_pred=lr.predict(X_test_norm)
```

```python
accuracy_score(y_test,lr_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(X_train_norm,y_train)
```

```python
svm_pred=svm.predict(X_test_norm)
```

```python
accuracy_score(y_test,svm_pred)
```
